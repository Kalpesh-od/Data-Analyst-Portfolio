```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2019,12,1)
end = datetime.datetime(2021,6,30)
```

# Gather Stock Information

```python
#STOCK INFORMATION ON SELECT FIRMS IN THE CONSUMER CYCLICAL INDUSTRY BASED ON MARKET CAP

# Unilever PLC
UL = data.DataReader('UL','yahoo',start,end)
# PepsiCo, Inc.
PEP = data.DataReader('PEP','yahoo',start,end)
# Kellogg Company
K = data.DataReader('K','yahoo', start,end)
# The Coca-Cola Company
KO = data.DataReader('KO', 'yahoo', start,end)
# Nestle S.A
NSRGF = data.DataReader('NSRGF','yahoo',start,end)
# Heineken N.V.
HINKF = data.DataReader('HINKF','yahoo', start,end)
# The Hershey Company
HSY = data.DataReader('HSY','yahoo', start,end)
# Conagra Brands, Inc.
CAG = data.DataReader('CAG','yahoo', start,end)
#Tyson Foods, Inc. (TSN)
TSN = data.DataReader('TSN','yahoo', start,end)
#The Kraft Heinz Company (KHC)
KHC = data.DataReader('KHC','yahoo', start,end)
#Suntory Beverage & Food Limited (STBFY)
STBFY = data.DataReader('STBFY','yahoo', start,end)
#Anheuser-Busch InBev SA/NV (BUD)
BUD = data.DataReader('BUD','yahoo', start,end)
#Mondelez International, Inc. (MDLZ)
MDLZ = data.DataReader('MDLZ','yahoo', start,end)
#Campbell Soup Company (CPB)
CPB = data.DataReader('CPB','yahoo', start,end)
#Ajinomoto Co., Inc. (AJINY)
AJINY = data.DataReader('AJINY','yahoo', start,end)
```

```python
tickers = ['UL','PEP','K','KO','NSRGF','HINKF', 'HSY','CAG','TSN','KHC','STBFY','BUD','MDLZ','CPB','AJINY'] 
```

```python
consumer_stocks = pd.concat([UL,PEP,K,KO,NSRGF,HINKF, HSY,CAG,TSN,KHC,STBFY,BUD,MDLZ,CPB,AJINY], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Consumer Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Consumer_Stock_Food 2.csv')
```
